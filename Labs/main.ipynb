### Task F

```python
college["Elite"] = pd.cut(college["Top10perc"], [0,50,100], labels=["No", "Yes"])
college["Elite"].value_counts()
sns.boxplot(data=college, x="Elite", y="Outstate")
```

### Task G

```python
# Börja här!
fig, axes = plt.subplots(2,2)
college["Apps"].plot.hist(bins=20, ax=axes[0,0])
axes[0,0].set_title("Apps")

college["Top10perc"].plot.hist(bins=50, ax=axes[0,1])
axes[0,1].set_title("Top10perc")

college["Enroll"].plot.hist(bins=70, ax=axes[1,0])
axes[1,0].set_title("Enroll")

college["Accept"].plot.hist(bins=10, ax=axes[1,1])
axes[1,1].set_title("Accept")
```

### Task H

```python
college_new = college.copy()
college_new["Elite"].replace(["Yes", "No"], [0,1], inplace=True)

x_yes = college_new[college_new["Elite"] == 0]["Personal"]
y_yes = college_new[college_new["Elite"] == 0]["Expend"]
x_no = college_new[college_new["Elite"] == 1]["Personal"]
y_no = college_new[college_new["Elite"] == 1]["Expend"]

plt.scatter(x=x_yes, y=y_yes, c="red", label="Elite")
plt.scatter(x=x_no, y=y_no, c="blue", label="Usual")

plt.xlabel("Personal spending")
plt.ylabel("Instructional spending")
plt.legend()
plt.show()
```

Det jag sökte efter var att kolla om det finns några samband mellan personalkostanader och undervisningskostander beroende om det var en "Usual" skola eller en "Elite" skola. 

Det jag kom fram till med hjälp av denna scatterplot är att i flesta fall så är det elite skolorna som har dyrare undervisningskostander. Som man kan se på scattern så är det enbart 1 blå prick som är högre upp när det kommer till utbislningskostander vilket det är betydligt mer av röd som motsvarar elitskolorna. 

Det som är lite intressamt här är att när man kollar på personalkostnader så ser man en tydlig trend att "vanliga" skolor är dyarare när det kommer till personalkostnader. Man kan ju dra en slutsats att det borde vara för att det går mer elever på "valiga" skolor i jämförelse med elite skolor och därav stiger personalkostanderna då skolorna behöver mer personal.
